```python
#importing dataset
dataset = pd.read_csv("Iris.csv")
```

```python
dataset.head(5)
```

```python
dataset.info()
```

```python
#checking for missing values
dataset.isna().sum()
```

```python
#converting categorical variables into numeric
dataset["Species"] = dataset["Species"].astype('category')
dataset["Species"] = dataset["Species"].cat.codes
```

```python
dataset.head(5)
```

```python
#classifying independent and dependent variables
x = dataset.iloc[:,1:5].values
y = dataset.iloc[:,-1].values
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#training the model
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
#prediction
y_pred = dtree.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
feat = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
cls = ['setosa','versicolor','virginica']
```

```python
#vizualising the decision tree
fig = plt.figure(figsize=(25,20))
Viztree = tree.plot_tree(dtree, feature_names=feat, class_names=cls, filled=True)
```
